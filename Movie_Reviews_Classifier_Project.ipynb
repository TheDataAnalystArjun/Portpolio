```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import nltk
import re
import string
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv('/content/IMDB Dataset.csv.zip')
df.head()
```

```python
df['sentiment'].unique()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
corpus = []
for i in range(0, len(df)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', df['review'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
df['review'] = corpus
```

```python
#tf idf
tf_idf = TfidfVectorizer()
#applying tf idf to training data
x = tf_idf.fit_transform(df['review'])

x = tf_idf.transform(df['review'])
```

```python
#Encoding
mappings = {}

mappings['sentiment'] = dict(zip(df['sentiment'].unique(), range(len(df['sentiment'].unique()))))
df['sentiment'] = df['sentiment'].map(lambda x: mappings['sentiment'][x])
df
```
